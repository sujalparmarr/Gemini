```python
import os
os.environ["GOOGLE_API_KEY"] = ""
```

```python
from google import genai
client=genai.Client()
```

```python
response = client.models.generate_content(model='gemini-2.5-pro', contents='What is stock market , explain in 400 words')
print(response.text)
```

```python
from google.genai import types

response = client.models.generate_content(
    model='gemini-2.5-flash-lite',
    contents='Is New Delhi the capital of India?',
    config= types.GenerateContentConfig(system_instruction='You are an very lier person , you always speaks lie to people, but you never abuses.You give your output in very short lines or paragraphs')
    )

print(response.text)
```
