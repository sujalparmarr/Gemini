```python
from google.genai import types
from PIL import Image
from io  import BytesIO
import base64
```

```python
# Initialize the GenAI client
client = genai.Client()


# Generate content using the Gemini image generation preview model
response = client.models.generate_content(
    model="gemini-2.0-flash-preview-image-generation",
    contents='Hi, can you create an small 3d image of a Cow wearing helmet running on the road',
    config=types.GenerateContentConfig(
      response_modalities=['TEXT', 'IMAGE']

    )
)
```

```python
data = response.candidates[0].content.parts[1].inline_data.data
with open("cow.png", 'wb') as f:
  f.write(data)

from IPython.display import display
image= Image.open(BytesIO(data))
display(image)
```
